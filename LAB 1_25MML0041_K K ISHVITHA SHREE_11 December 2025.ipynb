```python
import tensorflow as tf
import numpy as np

# I have used CIFAR dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()

all_images = np.concatenate((x_train, x_test), axis=0)

print(f"Total images loaded: {len(all_images)}")
print(f"Shape of the first image: {all_images[0].shape}")

first_image_shape = all_images[0].shape
all_have_same_shape = True

# A new list is needed to store the resized images
processed_images = []

for image in all_images:
    if image.shape != first_image_shape:
        all_have_same_shape = False
        print(f"Found image with different shape: {image.shape}. We reshape it")
        
        resized_image_tensor = tf.image.resize(image, (first_image_shape[0], first_image_shape[1]))
        
        reshaped_image = resized_image_tensor.numpy().astype(np.uint8)
        processed_images.append(reshaped_image)
    else:
        processed_images.append(image)
    


all_images = np.array(processed_images)


if all_have_same_shape:
    print(f"\nResult: All images naturally had the same shape of {first_image_shape}.")
else:
    print(f"\nResult: Images were inconsistent; all have now been reshaped to {all_images[0].shape}.")
```

Steps done:
- I have loaded CIFAR dataset
- Then concatenated train and test in all_images one
- I identified shape of the image
- for loop initialized, if any image has different shape then return False and Found image with different shape
- Since in this dataset I have same shape for all images I  have printed the final Result that all images have same shape
- If i found any image in different shape, then make all images in same shape as the first 
- Check the First Image's Shape This shape (32x32 pixels, 3 channels) becomes the target shape.
- An empty list called processed_images is created to store the images after they have been checked or modified.
- The code starts reviewing each images.
- Inside the loop, it checks if the current image's shape matches the first image's target shape.

- If an image has a different shape,it is noted that the dataset is inconsistent (all_have_same_shape = False).
- It then uses TensorFlow to stretch or shrink that image to exactly match the target shape.
- The newly reshaped image is added to the processed_images list.
- If an image already has the correct shape, it is simply added to the processed_images list without modification.
- The list of processed images is converted back into a single, unified NumPy array.
- The program prints a message confirming that, after the loop finishes, all images in the final dataset now have an identical shape.
